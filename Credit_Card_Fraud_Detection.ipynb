# Credit Card Fraud Detection Using Machine Learning

## l.Introduction
The credit card fraud detection project using Machine Learning aims to automatically identify fraudulent transactions by analyzing patterns and behaviors in transaction data. Machine learning algorithms learn from historical data to detect anomalies and predict the likelihood of fraud in real-time. This project improves accuracy, reduces false alarms, and enhances security in financial systems by making fraud detection faster and smarter.

## ll.Objective
To develop a classification model that accurately detects whether a credit card transaction is fraudulent or legitimate, using transaction data and machine learning techniques.

## Vl. Under-Sampling

#### Build a sample dataset containing similar distribution of Normal Transactions and Fraudulent Transactions

#### Number of Fraudulent Tansaction -->492

```python
legit_sample = legit.sample(n=492)
```

#### Concatenating Two DataFrames

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

## Vll. Splitting the data into Features & Targets

```python
X = new_dataset.drop(columns ='Class',axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

## Vlll. Split the data into Training data & Testing data

```python
X_train , X_test, Y_train,Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
# Successfully Split our data
```

## lX. Model Training & Evaluation

### Logistic Regression

```python
model = LogisticRegression()
```

#### Training the Logistic Regression Model with Training data

```python
model.fit(X_train, Y_train)
```

#### Accuracy on training data

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction , Y_train)
```

```python
X_pred_lr = model.predict(X_train)
```

```python
score_lr = round(accuracy_score(X_pred_lr,Y_train)*100,2)

print("The accuracy score achieved using Logistic Regression in Test Data is: "+str(score_lr)+" %")
```

#### Accuracy on test data

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
Y_pred_lr = model.predict(X_test)
```

```python
score_lr = round(accuracy_score(Y_pred_lr,Y_test)*100,2)

print("The accuracy score achieved using Logistic Regression in Test Data is: "+str(score_lr)+" %")
```
